```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
```

```python
data = pd.read_csv('Downloads/canada_per_capita_income.csv')
data
```

```python
#Converting the data to python DataFrame

df = pd.DataFrame(data)
```

```python
plt.plot(df['year'],df['per capita income (US$)'])
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
```

```python
plt.scatter(df['year'],df['per capita income (US$)'],color='red')
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
```

```python
# Assigning object for my model
canada = linear_model.LinearRegression()

# Fitting the model
canada.fit(df[['year']],df['per capita income (US$)'])
```

```python
plt.scatter(df['year'], df['per capita income (US$)'],color='red', marker='+')
plt.xlabel('year', fontsize=15)
plt.ylabel('per capita income (US$)', fontsize=15)
plt.title('Linear Regression', fontsize=15)
plt.plot(df['year'], canada.predict(df[['year']]))
```

```python
canada.predict(pd.DataFrame({'year': [2020, 2021, 2022, 2023, 2024, 2025, 2026]}))
```
